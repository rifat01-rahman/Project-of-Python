```python
from matplotlib.image import imread
from matplotlib import pyplot as plt
import os
```

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("abrahamanderson/ladybug")

print("Path to dataset files:", path)
```

```python
import os

print(os.listdir(path))
```

```python
from matplotlib.image import imread
import os

image = imread(os.path.join(path, "ladybug-daisy.webp"))
print(image.shape)
```

```python
from sklearn.cluster import KMeans
```

```python
X = image.reshape(-1, 3)
kmeans = KMeans(n_clusters=8).fit(X)
segmented_img = kmeans.cluster_centers_[kmeans.labels_]
segmented_img = segmented_img.reshape(image.shape)
```

```python
plt.figure(figsize=(10,4))

# Original image
plt.subplot(1,2,1)
plt.imshow(image)
plt.title("Original")
plt.axis('off')

# Segmented image
plt.subplot(1,2,2)
plt.imshow(segmented_img.astype('uint8'))
plt.title("Segmented (8 colors)")
plt.axis('off')

plt.show()
```
